# Merge with regimen table
First open the regimen table, then select those rows where MERGED_TUMOUR_ID is in the breast_cancer_patients, then make a new table with ID, Primary Diagnosis, and the Regimen

```python
sact_regi_data = pd.read_csv('simulacrum/data/sim_sact_regimen.csv')
```

```python
df = sact_regi_data.merge(breast_cancer_patients, on='MERGED_TUMOUR_ID')
```
